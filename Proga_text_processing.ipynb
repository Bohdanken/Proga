```python
from datetime import datetime
from bs4 import BeautifulSoup
import glob
import pandas as pd
BASE_URL = "https://www.understandingwar.org/backgrounder/russian-offensive-campaign-assessment"
INPUT_DATA_FOLDER ="D:/Proga/fff"

UTPUT_FOLDER ="D:\\"
OUTPUT_DATA_FILE = "cleared_ reports. csv"

Files_by_days = glob.glob(f"{INPUT_DATA_FOLDER}/*.html")

all_data = []

for file in Files_by_days:
    d={}
    file_name = file.split("\\")[-1].replace(".html","")
    data_info=file_name.split("-")
    date=datetime.strptime(data_info[0]+"/"+data_info[1]+"/"+data_info[2],"%Y/%m/%d")
    url=""
    with open(file, "r",encoding='utf-8') as cfile:
        #url=cfile.readline()
        parsed_html= BeautifulSoup(cfile.read())
        title = parsed_html.head. find('title').text
        textS_title = parsed_html.body.find('h1', attrs={"id":'page-title'}).text
        text_main = parsed_html.body.find('div', attrs={'class': 'field field-name-body field-type-text-with-summary field-label-hidden'})
        d={"date":date,
           #"url":url,
           "title": title,
           "text_title":text_title,
           "text_main":text_main}
        all_data.append(d)
```

```python
df = pd.DataFrame.from_dict(all_data)
```

```python
 df = df.sort_values (by=["date"])
```
